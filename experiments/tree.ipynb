```python
import random
import math

class Node:
    def __init__(self, state):
        self.state = state
        self.visits = 0
        self.wins = 0
        self.children = []

    def expand(self):
        # Generate all possible child states
        child_states = self.state.generate_child_states()

        # Create child nodes for each child state
        for child_state in child_states:
            child_node = Node(child_state)
            self.children.append(child_node)

    def select_child(self, exploration_constant):
        # Select the child node with the highest UCB1 value
        best_child = None
        best_ucb1 = float('-inf')

        for child in self.children:
            ucb1 = child.wins / child.visits + exploration_constant * math.sqrt(math.log(self.visits) / child.visits)
            if ucb1 > best_ucb1:
                best_ucb1 = ucb1
                best_child = child

        return best_child

    def simulate(self):
        # Simulate a random playout from the current state
        current_state = self.state

        while not current_state.is_terminal():
            possible_moves = current_state.get_possible_moves()
            random_move = random.choice(possible_moves)
            current_state = current_state.make_move(random_move)

        return current_state.get_result()

    def backpropagate(self, result):
        # Update the visits and wins count of the current node and its ancestors
        self.visits += 1
        self.wins += result

        if self.parent is not None:
            self.parent.backpropagate(result)

    def uct_search(self, num_iterations, exploration_constant):
        for _ in range(num_iterations):
            node = self

            # Selection
            while not node.children and not node.state.is_terminal():
                node.expand()
                node = node.select_child(exploration_constant)

            # Expansion
            if node.children:
                node = random.choice(node.children)

            # Simulation
            result = node.simulate()

            # Backpropagation
            node.backpropagate(result)

        # Select the best move based on the number of visits
        best_move = None
        best_visits = float('-inf')

        for child in self.children:
            if child.visits > best_visits:
                best_visits = child.visits
                best_move = child.state.get_last_move()

        return best_move
```
